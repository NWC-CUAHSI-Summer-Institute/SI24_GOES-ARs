```python
import xarray as xr
import os
import netCDF4 as nc
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from bs4 import BeautifulSoup
import requests
from tqdm import tqdm
```

## Download training data

```python
# URL of the directory containing .nc files
url = 'https://portal.nersc.gov/project/ClimateNet/climatenet_new/train/'

# Make a request to the website
response = requests.get(url)
response.raise_for_status()  # Ensure the request was successful

# Parse the content of the page
soup = BeautifulSoup(response.text, 'html.parser')

# Find all links on the page
links = soup.find_all('a')

for link in tqdm(links):
  # Check if the link's href attribute ends with '.nc'
  if link['href'].endswith('.nc'):
    file_url = url + link['href']  # Complete the file URL
    file_name = link['href']
    # print(f"Downloading {file_name}...")

    if not os.path.exists(file_name):
      # Download the file
      response = requests.get(file_url)
      response.raise_for_status()  # Ensure the file request was successful

      # Save the file
      with open(file_name, 'wb') as f:
          f.write(response.content)

      # print(f"Downloaded {file_name}")
print("All files downloaded.")
```
